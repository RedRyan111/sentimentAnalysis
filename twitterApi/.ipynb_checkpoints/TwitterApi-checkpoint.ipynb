```python
tweet_dict = {}
```

```python
for tweet in tweepy.Cursor(api.search, q = searchQuery, count = 10, lang ="en").items(max_tweets):
        ID = tweet.id
        txt = tweet.text
        tweet_dict[ID]=txt
```
